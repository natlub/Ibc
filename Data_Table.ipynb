```python
from c0m_func import *
import pandas
from datetime import datetime
```

```python
pathlist, snlist, folderlist, filelist, instlist, filtlist, datelist, exposurelist = scan_data("/Users/natalie/Documents/sn/*/*/*c0m.fits")  
# data of sn's in directory - also checks **CORR commends, does not count Omit because it means it is not required
```

```python
# scan matching sn position and explosion date
ra_h, ra_m, ra_s, dec_d, dec_m, dec_s, explist = csvscan(snlist, "/Users/natalie/Documents/c0mdata.csv")  
```

```python
# create ra sec sn's table
ra_dec_table = Table([snlist, ra_h, ra_m, ra_s, dec_d, dec_m, dec_s], names=('sn', 'ra_h', 'ra_m', 'ra_s', 'dec_d', 'dec_m', 'dec_s'))  
```

```python
# convert ra dec to deg
ra_deg = ra2deg(ra_dec_table) 
dec_deg = dec2deg(ra_dec_table)
```

```python
# create the general table
data_table = Table([snlist, pathlist, folderlist, filelist, instlist, filtlist, exposurelist, ra_deg, dec_deg, datelist, explist], names=('sn', 'path', 'folder', 'file', 'inst', 'filt', 'exposure', 'ra_deg', 'dec_deg', 'imgdate', 'expdate'))
```

```python
# taking only WFPC2 cam
rows = []
for n in range(0, len(data_table)):
    if data_table[n]['inst'] != 'WFPC2':
        rows.append(n)
data_table.remove_rows(rows)
```

```python
# finding the best det, and the px of the ra dec, that are with distance of buffer from the edges - if not found within buffer - 0 is listed. tried 80 - only 2 zeroes
det, x_px, y_px = find_det_and_px(data_table, 5)
```

```python
# adding the lists above to the data_table
det_col = Column(name='det', data=det)
data_table.add_column(det_col)
x_px_col = Column(name='x_px', data=x_px)
data_table.add_column(x_px_col)
y_px_col = Column(name='y_px', data=y_px)
data_table.add_column(y_px_col)
```

```python
# change to a jupyter table
dt=data_table.to_pandas()
dt.to_pickle('data.pickle')
df=pandas.read_pickle('data.pickle')
```

```python
# taking only img pre-exp
data_table_pre=data_table
rows = []
for n in range(0, len(data_table_pre)):
    imgd = datetime.strptime(data_table_pre[n]['imgdate'], "%Y-%m-%d")
    expd = datetime.strptime(data_table_pre[n]['expdate'], "%m/%d/%Y")
    if imgd > expd:
        rows.append(n)
data_table_pre.remove_rows(rows)
```

```python
# change to a jupyter table
dt=data_table_pre.to_pandas()
dt.to_pickle('data.pickle')
df_pre=pandas.read_pickle('data.pickle')
```

```python
# taking only img post-exp
data_table_post=data_table
rows = []
for n in range(0, len(data_table_post)):
    imgd = datetime.strptime(data_table_post[n]['imgdate'], "%Y-%m-%d")
    expd = datetime.strptime(data_table_post[n]['expdate'], "%m/%d/%Y")
    if imgd < expd:
        rows.append(n)
data_table_post.remove_rows(rows)
```

```python
# change to a jupyter table
dt=data_table_post.to_pandas()
dt.to_pickle('data.pickle')
df_post=pandas.read_pickle('data.pickle')
```
